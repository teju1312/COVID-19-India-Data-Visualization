```python
# 2.remove duplicates nd check
df = pd.read_csv("state_level_latest.csv")
df

df[df.duplicated()]
```

```python
#3. change type of date to datetime
df = pd.read_csv("state_level_latest.csv")
df['Last_Updated_Time'] = pd.to_datetime(df['Last_Updated_Time'],format='mixed')


df.info()
```

```python
# 4.remove unwanted columns
df = pd.read_csv("state_level_latest.csv")
df = df.drop(columns=['Migrated_Other','State_Notes','Delta_Confirmed','Delta_Recovered','Delta_Deaths'])
df
```

```python
# 5.check no of uniquess
df.nunique()
```

```python
# 6.keep "total' row at last
# Separate the 'Total' row
total_row = df[df['State'] == 'Total']
other_rows = df[df['State'] != 'Total']

# Concatenate the other rows with the total row at the end
df= pd.concat([other_rows, total_row])

df
```

```python
# # .check for no duplicates
df.duplicated().sum()
```

```python
# 7.re-load into excel
# firstly import openpyxl module in environment of anaconda navigator
import openpyxl
# to excel
df.to_excel("state_level_cleaned.xlsx", index=True)

# to csv
df.to_csv("state_level_cleaned.csv",index = True)
```
