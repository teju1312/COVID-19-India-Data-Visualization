```python
# 2.remove unwanted columns
df = pd.read_csv("india_non-virus-deaths.csv")
df = df.drop(columns=['DB_LASTUPDATED_SET','Source','Date of publication','Link','Category'])
df
```

```python
#3. check count of unique

df.nunique()
```

```python
# 4.change types
df['No. of deaths'] = df['No. of deaths'].astype("Int64")
df['S. No'] =df['S. No'].astype("Int64")
df
df.info()
df.nunique()
```

```python
# 5.set index
df = df.set_index('State/UT')
df
```

```python
# 6.sort values
df = df.sort_index()
df
```

```python
# 7.checking duplicates
df[df.duplicated()]
```

```python
# # .check for no duplicates
df.duplicated().sum()

df.isnull()
```
