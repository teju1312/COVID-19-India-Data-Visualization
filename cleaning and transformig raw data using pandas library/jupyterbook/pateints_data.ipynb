```python
#1. check types and nulls

pd.read_csv("patients_data.csv",low_memory=False).info()
```

```python
# 2.remove empty columns
df = pd.read_csv("patients_data.csv",low_memory=False)
df

# check
df["Estimated Onset Date"].isnull()

# remove
df = df.drop(columns=['Estimated Onset Date'])
df
```

```python
# 3.remove unneccessary columns
df = df.drop(columns=[
    'State Patient Number',
    'Notes',
    'Contracted from which Patient (Suspected)',
    'Nationality',
    'Type of transmission',
    'Source_1',
    'Source_2',
    'Source_3',
    'Backup Notes',
    'Entry_ID'
],errors='ignore')
df
```

```python
df.info()

```

```python
df.nunique()
```

```python
# 4.remove duplicates nd check

df = df.drop_duplicates()
df

df = df.drop_duplicates(subset=['Patient Number','Date Announced','Age Bracket','Gender','Detected City','Detected City','Detected State','State code','Current Status','Status Change Date','Num Cases'])
df.duplicated().sum()
```

```python
# 5.change type of patient no & num cases
df['Patient Number'] = df['Patient Number'].astype('Int64')
df['Num Cases'] = df['Num Cases'].astype('Int64')


# 6. make index 
df = df.set_index('Patient Number')
df

```

```python
#7. sort data 
df = df.sort_index()
df


# same as 
df = df.sort_values(by='Patient Number')
df
```

```python
df.info()
```

```python
#8. change type of date to datetime 
df['Date Announced'] = pd.to_datetime(df['Date Announced'],format='mixed')
df['Status Change Date'] = pd.to_datetime(df['Status Change Date'],format='mixed')


# # another way
# df = pd.read_csv("patients_data.csv",low_memory=False,parse_dates=["Date Announced"],dayfirst=True)
# df = pd.read_csv("patients_data.csv",low_memory=False,parse_dates=["Status Change Date"],dayfirst=True)

df.info()
```
