```python
#3. change type of date to datetime and set date as index
df=pd.read_csv("covid-hospitalizations.csv")
df['date'] = pd.to_datetime(df['date'],format="%Y-%m-%d")


# another way
df=pd.read_csv("covid-hospitalizations.csv",parse_dates=['date'])
df.info()

#### date as index
df = df.set_index('date')
df
```

```python
# 4. column's case convertion
df.columns = df.columns.str.title()
df
```

```python
# 5. set index's case
df.index.name = "Date"
df
```

```python
#6. sort data by date
df = df.sort_index()
df

# same as 
df = df.sort_values(by='Date')
df
```
